```python
import pandas as pd

data="""Name,Math,Science,English
Amit,78,85,74
Priya,92,89,96
Kiran,65,70,60
Anita,88,95,90
Ravi,55,60,58"""

with open("/content/drive/MyDrive/Colab Notebooks/students.csv","w") as f:
  f.write(data)
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/students.csv")
print(df.head(3))
```

2.Calculate Average Scores

Add a new column Average that calculates the mean score for Math, Science, and
English for each student.
**bold text**
Print the student with the highest average.

```python
df['Average']=df[['English','Math','Science']].mean(axis=1)
high_avg=df[df['Average']==df['Average'].max()]
print("Highest Avg Student\n",high_avg)
```

3. Filter Students

Display all students who have scored more than 80 in Science.

Display all students with an average score below 70.

```python
Sci_above_80=df[df['Science']>80]
print("Students with Science above 80\n",Sci_above_80)
```

```python
avg_below_70=df[df['Average']<70]
print("Students with Average below 70\n",avg_below_70)
```

4. Sorting

Sort the students by English score (descending) and print the top 3.

```python
sort_by_eng_score=df.sort_values(by='English',ascending=False)
print(sort_by_eng_score.head(3))
```

5. Chart Exercise

Create a bar chart showing students’ names on the X-axis and their average
scores on the Y-axis.

Create a line chart for all students’ scores in Math, Science, and English (use
df.plot() ).

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3))
plt.bar(df['Name'],df['Average'],color=['Red','Green'])
plt.xlabel('Students')
plt.ylabel('Average Scores')
plt.title('Students Average Scores')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
df.plot(x='Name',y=['Math','Science','English'],kind='line',color=['Red','Green','Yellow'])
plt.xlabel('Students')
plt.ylabel('Scores')
plt.title('Students Scores')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

6. Bonus Task

Find the subject with the highest total score across all students.

Save the updated DataFrame (with Average column) as a new CSV named
students_summary.csv .

```python
total_score=df[['Math','Science','English']].sum()
top_sub=total_score.idxmax()
print("Top Subject Mark\n",top_sub,total_score.max())
```
